### E1-3-1 귀하께서 현재 하시는 일은 정부 지원 일자리(노인일자리, 공공근로 등) 입니까? (예:1 아니오:2 비해당:9)

```python
df20.shape
```

```python
fig, ax = plt.subplots(1, 2, figsize=(14,7))

# Bar chart
values = df20['E1_3_1'].value_counts().sort_index()
bars = values.plot(kind='bar', ax=ax[0])

ax[0].set_title('Bar Chart')
ax[0].set_xlabel('Value')
ax[0].set_ylabel('Frequency')

# Add labels on top of the bars
for bar in bars.patches:
    bars.annotate(format(bar.get_height(), '.0f'),
                   (bar.get_x() + bar.get_width() / 2, 
                    bar.get_height()), 
                   ha='center', 
                   va='center',
                   size=10, 
                   xytext=(0, 8),
                   textcoords='offset points')

# Pie chart
df20['E1_3_1'].value_counts().sort_index().plot(kind='pie', autopct='%1.1f%%', ax=ax[1])
ax[1].set_title('Pie Chart')
ax[1].set_ylabel('')  # to remove the 'None' ylabel

plt.show()
```

설문조사 참여자 중 정부지원 일자리 참여자는 약 800명입니다. Causal forest를 돌리기에 충분한 숫자인지는 잘 모르겠네요.

설문조사 답변자 중 근로소득이 있는 분들의 수도 궁금해져서 돌려봤습니다.

```python
df_filtered = df20[df20['E1_3_1'] == 1]

plt.figure(figsize=(10,6))

sns.barplot(x=df_filtered['E1_3'], y=df_filtered['J3b_1_1'], ci=95)
plt.title('Confidence Intervals for yearly labour income grouped by E_1_3')
plt.xlabel('E1_3')
plt.ylabel('J3b_1_1')

plt.show()
```

근로시간과 시급도 살펴봤습니다

```python
df_filtered = df20[df20['E1_3_1'] == 1]

plt.figure(figsize=(14, 6))

# First subplot for weekly working hours
plt.subplot(1, 2, 1)
sns.barplot(x=df_filtered['E1_3'], y=df_filtered['E1_5_2'], ci=95)
plt.title('Confidence Intervals for weekly working hours grouped by E_1_3')
plt.xlabel('E1_3')
plt.ylabel('J3b_1_1')

# Second subplot for hourly wage
plt.subplot(1, 2, 2)
sns.barplot(x=df_filtered['E1_3'], y=df_filtered['J3b_1_1'] / df_filtered['E1_5_2'] / 52, ci=95)
plt.title('Confidence Intervals for hourly wage grouped by E_1_3')
plt.xlabel('E1_3')
plt.ylabel('J3b_1_1')

plt.tight_layout()

plt.show()
```

확실히 상용근로자가 근로시간도 길고, 시급도 훨씬 높습니다. 저번에 말씀드렸다시피, 윤정부가 공공형 일자리를 줄이고 민간형을 늘린다고 했는데, 참고할 만한 데이터 같네요.
