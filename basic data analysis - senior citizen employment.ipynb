## 2.1 근로소득 금액에 관한 분석

### 첫째로는 설문조사 응답자 중 정부지원 일자리에 참여하고 있는 사람들의 도표입니다.

```python
fig, ax = plt.subplots(1, 2, figsize=(14,7))

# Bar chart
values = df20['E1_3_1'].value_counts().sort_index()
bars = values.plot(kind='bar', ax=ax[0])

ax[0].set_title('Bar Chart')
ax[0].set_xlabel('Value')
ax[0].set_ylabel('Frequency')

# Add labels on top of the bars
for bar in bars.patches:
    bars.annotate(format(bar.get_height(), '.0f'),
                   (bar.get_x() + bar.get_width() / 2, 
                    bar.get_height()), 
                   ha='center', 
                   va='center',
                   size=10, 
                   xytext=(0, 8),
                   textcoords='offset points')

# Pie chart
df20['E1_3_1'].value_counts().sort_index().plot(kind='pie', autopct='%1.1f%%', ax=ax[1])
ax[1].set_title('Pie Chart')
ax[1].set_ylabel('')  # to remove the 'None' ylabel

plt.show()
```

다만, 분석의 허점이 하나 있습니다: 이 분석은 아래와 같은 방식으로 진행되었습니다:

- 노인일자리 사업에 참여하고 있는 분들을 추려낸다.
- 추려낸 분들을 근로형태 (상용, 임시, 일용)으로 그룹화한다.
- 그룹에 따른 소득 분포를 그린다.

여기서 문제가, 이 분들의 소득이 전부 노인일자리사업에서 오는 돈은 아니라는 것입니다. 다른 사이드잡이 있을수도 있죠. 몇시간 공익활동만 하는 분들의 경우 이 가능성은 더더욱 높고요.

현재 노인일자리사업에 참여 중인 분들의 참여의향은 78%로, 과거 참여했던 분들의 재참여 의향인 40%보다 훨씬 높았습니다. 

현재 노인일자리사업이 점차 개선되어서인지, 아니면 이전에 참여했던 분들은 과거와 달리 체력적 여건이 안되는 분들이 많아서인지, 경제적으로 자유로워져서인지, 여러가지 가능성이 있겠지만, 잘 모르겠군요.

### 넷째로, 현재 참여 중인 분들 중, 근로형태에 따른 참여의향을 살펴보았습니다.

E1_3: 1 상용 근로자 | 2 임시 근로자 | 3 일용 근로자

```python
# Filter the dataframe based on the condition df20['E1_3_1'] == 1
df_filtered = df20[df20['E1_3_1'] == 1]

# Group the filtered dataframe by E1_3 and E4 columns and count occurrences
grouped_df = df_filtered.groupby(['E1_3', 'E5']).size().unstack(fill_value=0)

# Calculate the percentage
grouped_df_percentage = grouped_df.apply(lambda x: x / x.sum() * 100, axis=1)

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot the count bar chart
grouped_df.plot(kind='bar', stacked=False, ax=axes[0])
axes[0].set_title('Count of E5 Grouped by E1_3 (Filtered by E1_3_1 == 1)')
axes[0].set_xlabel('E1_3')
axes[0].set_ylabel('Count')
axes[0].legend(title='E5', title_fontsize='12')
axes[0].tick_params(axis='x', rotation=0)

# Plot the percentage bar chart
grouped_df_percentage.plot(kind='bar', stacked=False, ax=axes[1])
axes[1].set_title('Percentage of E5 Grouped by E1_3 (Filtered by E1_3_1 == 1)')
axes[1].set_xlabel('E1_3')
axes[1].set_ylabel('Percentage')
axes[1].legend(title='E5', title_fontsize='12')
axes[1].tick_params(axis='x', rotation=0)

# Adjust the layout
plt.tight_layout()

# Display the plot
plt.show()
```

*현재 노인일자리 사업에 참여하고 있는 사람들의 유형별 참여의향 비율 (1 상용 근로자 | 2 임시 근로자 | 3 일용 근로자)*

1 상용 근로자 | 2 임시 근로자 | 3 일용 근로자

예상과 달리, 상용근로자가 임시/일용근로자보다 참여의향이 낮게 나옵니다. 다만 샘플사이즈가 적은 건 감안할 필요가 있어보입니다.
